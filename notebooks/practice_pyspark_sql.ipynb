### Your Practice Query Below
Try a complex query, like finding the top 5 nations by total order volume:

```python
%%sql
SELECT 
    n.n_name AS nation,
    SUM(o.o_totalprice) AS total_revenue
FROM orders o
JOIN customer c ON o.o_custkey = c.c_custkey
JOIN nation n ON c.c_nationkey = n.n_nationkey
GROUP BY 1
ORDER BY 2 DESC
LIMIT 5;
```
